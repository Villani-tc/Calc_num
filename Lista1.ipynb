O exercício trata-se do teorema de Monte Carlo, no qual há um circulo e tenho vários valores, entre 0 e 1, pois estamos no primeiro quadrante, porem, haverá pontos que, se distribuidos aleatoriamente, sairão do quadrante, que serão nossos erros, e devem ser mensurados, para melhor aproximação do número de pi.

Ao resolver o exercicio, podemos seguir por dois caminhos, com a separação logaritimica da sequencia, ou, da separação linear.

# Forma Logarítmica

No ínicio, por se tratar de uma função logaritima, os primeiros pontos terão altos erros, por serem menos pontos, e, no final, a curva de erro se aproximará de 0

Quanto mais amostras eu tiver, ou seja, mais bolinhas jogar, maior a probabilidade de ser exato

```python
import numpy as np
import matplotlib.pyplot as plt

# Número de pontos amostrados
n_log = np.logspace(1, 6, num=100, dtype=int)  # De 10 a 1.000.000 pontos
n_log
```

```python
errors = [] #criar uma lista para armazenar o valor de diferença do pi calculado para o pi real, estou olhando oq ta dentro do circulo
errors_rel = []
# Valor real de pi
pi = np.pi

for n in n_log:
    x = np.random.uniform(0, 1, n) #o n será os pontos que geramos acima, de forma logaritmica
    y = np.random.uniform(0, 1, n)

    m = np.sum(x**2 + y**2 <= 1)  # equação de circunferência de raio unitario (x^2 + y^2 = 1) Se for 1+1 será 2, e estará fora do circulo! Aqui estou contando o que está dentro
    pi_estimate = 4 * m / n  # Estimativa de pi

    error = abs(pi_estimate - pi)
    errors.append(error)

    error_rel = abs(error/pi)
    errors_rel.append(error_rel)

# Gráfico da evolução do erro absoluto - escala logaritmica
plt.figure(figsize=(16, 9))
plt.loglog(n_log, errors, marker='o', linestyle='-', label='Erro absoluto')
plt.xlabel('Número de pontos (n)')
plt.ylabel('Erro absoluto')
plt.title('Evolução do erro absoluto na estimativa de π')
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()

# Gráfico da evolução do erro relativo - escala logaritmica

plt.figure(figsize=(16, 9))
plt.loglog(n_log, errors_rel, marker='o', linestyle='-', label='Erro relativo')
plt.xlabel('Número de pontos (n)')
plt.ylabel('Erro relativo')
plt.title('Evolução do erro relativo na estimativa de π')
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()

# Gráfico da evolução do erro absoluto + relativo - escala logaritmica
plt.figure(figsize=(16, 9))
plt.loglog(n_log, errors, marker='o', linestyle='-', label='Erro absoluto')
plt.loglog(n_log, errors_rel, marker='+', linestyle='-', label='Erro relativo')
plt.xlabel('Número de pontos (n)')
plt.ylabel('Erros')
plt.title('Evolução dos Erros na Estimativa de π')
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

### Metodo de Newton-Raphson

O Método de Newton–Raphson tem o objetivo de estimar as raízes de uma função. Para isso, escolhe-se uma aproximação inicial para esta. Após isso, calcula-se a equação da reta tangente (por meio da derivada) ao gráfico da função nesse ponto e a interseção dela com o eixo das abcissas, a fim de encontrar uma melhor aproximação para a raiz. Se repete esse processo várias vezes, encontrando as raízes mais rapidamente do que com métodos convencionais.

```python
import math as m
import matplotlib.pyplot as plt

# Dar valor às variáveis
l1 = 8
l2 = 10
v = (3 * m.pi) / 5


# Definir a função da equação não-linear
def f(a):
    y = l2 * (m.cos(m.pi - v - a) / m.sin(m.pi - v - a)**2) - l1 * (m.cos(a) / m.sin(a)**2)
    return y

# Definir a função do peso da barra
def l(a):
    L = l2 / m.sin(m.pi - v - a) + l1 / m.sin(a)
    return L

# Definir a derivada função da equação não-linear
def g(a):
    z = - l1 / m.sin(a) + 2*l1*m.cos(a)**2 / m.sin(a)**3 + l2 / m.sin(a + v) + 2*l2*m.cos(a + v)**2 / m.sin(a + v)**3
    return z

def newton_raphson(alpha0, t, tol=1e-6):
    alpha = alpha0
    vetornr = []

    for i in t:
        f_val = f(alpha)
        df_val = g(alpha)

        if abs(df_val) < 1e-8:
            print("Derivada próxima de zero, encerrando.")
            break

        alpha_new = alpha - f_val / df_val
        vetornr.append(l(alpha_new))

        print(f"Iteração {i}: α = {alpha_new:.6f} rad ({m.degrees(alpha_new):.6f} graus)")

        if abs(alpha_new - alpha) < tol:
            print("Convergência alcançada.")
            # Calcular o tamanho L da barra
            print(f"O tamanho L da barra é {round(l(alpha), 7)}")
            break

        alpha = alpha_new



    return vetornr

# Execução com vetor de iterações t
t = list(range(10))
alpha0 = m.pi / 4  # chute inicial
vetornr = newton_raphson(alpha0, t)



plt.plot(t, vetornr)
plt.grid()
plt.ylabel('Valores do tamanho L')
plt.xlabel('Numero n de aproximações')
plt.title('Função do tamanho (Newton-Raphson)')
plt.show()
```

### Método da Bisseção

O método da bisseção utiliza a ideia de que uma função contínua tem um zero no intervalo caso [a, b] se f(a) . f(b) < 0.

Para encontrar a raiz da função se repete esse processo inúmeras vezes, cada vez mais diminuindo o intervalo, alterando a ou b, dependendo de qual "quadrante" o zero se encontra, após dividido no meio.

```python
import math as m
import matplotlib.pyplot as plt

def f(a):
    y = l2 * (m.cos(m.pi - v - a) / (m.sin(m.pi - v - a)**2)) - l1 * (m.cos(a) / m.sin(a)**2)
    return y

# Definir o valor inicial (x0)
vetorbis = []
iteracoes = []

a = 0.575
b = 0.605
e = 0.0001

# Teorema de Bolzano
r = 1

if f(a) * f(b) < 0:
    while (m.fabs(b - a) / 2 > e):
        xi = (a + b) / 2

        vetorbis.append(l(xi))
        iteracoes.append(r)

        print(f"Para n={r}: {xi}")

        if f(xi) == 0:
            print("A raiz α é:", xi)
            break

        if f(a) * f(xi) < 0:
            b = xi
        else:
            a = xi

        r += 1

# Calcular o tamanho L da barra
print(f"O tamanho L da barra é {round(l(xi), 7)}")

# Gráfico da aproximação do valor do tamanho L
plt.plot(iteracoes, vetorbis)
plt.grid()
plt.ylabel('Valores do tamanho L')
plt.xlabel('Número n de aproximações')
plt.title('Função do tamanho (Bissecção)')
plt.show()
```

### Comparando os dois métodos

O método da bisseção é mais lento do que o método anterior, apresentando uma ordem de convergência linear de taxa 1/2, enquanto o método de Newton–Raphson possui uma ordem de convergência quadrática.

Portanto, o grupo conclui que ambos os métodos podem ser utilizados para descobrir o tamanho da barra, entretanto o método de Newton-Raphson consegue atingir o objetivo mais rapidamente, levando-nos a entender que seu uso é mais apropriado para essa situação.

```python
import numpy as np
import matplotlib.pyplot as plt
import math


def g_x(x, a):
  #Função iterativa do ponto fixo
  return x * (x**2 + 3*a) / (3*x**2 + a)

def g_prime_x(x, a):
 #Derivada da função iterativa do ponto fixo
  num = (3*x**2 + a)*(x**2 + 3*a) - x*(6*x*(3*x**2 + a) - 6*x*(x**2 + 3*a))
  den = (3*x**2 + a)**2
  return num / den

def verifica_convergencia(a, x_range):
  """Verifica se a condição |g'(x)| < 1 é atendida antes das iterações."""
  derivadas = [g_prime_x(x, a) for x in x_range]
  return all(abs(d) < 1 for d in derivadas)

def raiz_metodo_ponto_fixo(a, x0=1.0, tol=1e-10, max_iter=1000):
    """
    Calcula aproximadamente a raiz(a) usando a iteração de ponto fixo e retorna a evolução das aproximações
    """
    x = x0
    iteracoes = []
    valores_aproximados = []
    erros = []

    for k in range(1, max_iter+1):
      x_new = x * ((x**2 + 3*a) / (3*x**2 + a))
      iteracoes.append(k)
      valores_aproximados.append(x_new)
      erros.append(abs(x_new - math.sqrt(a)))

      if abs(x_new - x) < tol:
          return x_new, k, iteracoes, valores_aproximados, erros
      x = x_new
      print(f' Método do Ponto Fixo {x_new}')

    return x, max_iter, iteracoes, valores_aproximados, erros

def raiz_metodo_newton_raphson(a, x0=1.0, tol=1e-10, max_iter=1000):
    """
    Calcula aproximadamente a raiz(a) usando o método de Newton-Raphson
    """
    x = x0
    iteracoes = []
    valores_aproximados = []
    erros = []

    for k in range(1, max_iter+1):

      x_new = 0.5 * (x + (a / x))
      iteracoes.append(k)
      valores_aproximados.append(x_new)
      erros.append(abs(x_new - math.sqrt(a)))

      if abs(x_new - x) < tol:
          return x_new, k, iteracoes, valores_aproximados, erros
      x = x_new
      #print(f' Método de Newton Raphson {x_new}')

    return x, max_iter, iteracoes, valores_aproximados, erros

if __name__ == "__main__":

    x0 = 98
    a = 3891
    x_range = np.linspace(0.1 * math.sqrt(a), 10 * math.sqrt(a), 100)
    x_range_x0 = np.linspace(0.9 * x0, 1.1 * x0, 1)
    convergente = verifica_convergencia(a, x_range_x0)

    if convergente:
        print("O método de ponto fixo satisfaz a condição de convergência (|g'(x)| < 1).")
    else:
        print("O método de ponto fixo pode não convergir para a raiz desejada.")

    g_prime_vals = [g_prime_x(x, a) for x in x_range]

    x_aprox_pf, n_iter_pf, iteracoes_pf, valores_aproximados_pf, erros_pf = raiz_metodo_ponto_fixo(a, x0, tol=1e-6, max_iter=1000)
    x_aprox_nr, n_iter_nr, iteracoes_nr, valores_aproximados_nr, erros_nr = raiz_metodo_newton_raphson(a, x0, tol=1e-6, max_iter=1000)

    print("----------------------------------------------")
    print(f"     Método de Ponto Fixo:")
    print(f"     Aproximação da Raiz({a}): {x_aprox_pf}")
    print(f"     Número de iterações: {n_iter_pf}")
    print(f"     Erro absoluto: {abs(x_aprox_pf - math.sqrt(a))}")
    print("----------------------------------------------")
    print(f"     Método de Newton-Raphson:")
    print(f"     Aproximação da Raiz({a}): {x_aprox_nr}")
    print(f"     Número de iterações: {n_iter_nr}")
    print(f"     Erro absoluto: {abs(x_aprox_nr - math.sqrt(a))}")
    print("----------------------------------------------")


    plt.figure(figsize=(8,5))
    plt.plot(x_range, g_prime_vals, linestyle='-', color='blue', label="g'(x)")
    plt.axhline(y=1, color='r', linestyle='--', label="Limite de Convergência (|g'(x)| = 1)")
    plt.axhline(y=-1, color='r', linestyle='--')
    plt.xlabel('Valores de X')
    plt.ylabel("g'(x)")
    plt.title("Análise da Convergência do Método de Ponto Fixo")
    plt.legend()
    plt.grid()
    plt.show()

    plt.figure(figsize=(8,5))
    plt.plot(iteracoes_pf, valores_aproximados_pf, marker='o', linestyle='-', label='Ponto Fixo')
    plt.plot(iteracoes_nr, valores_aproximados_nr, marker='o', linestyle='-', label='NR')
    plt.axhline(y=math.sqrt(a), color='r', linestyle='--', label='Raiz Exata')
    plt.xlabel('Iterações')
    plt.ylabel('Valor Aproximado')
    plt.title('Evolução da Aproximação da Raiz Quadrada')
    plt.ylim(0, math.sqrt(a) *2)
    plt.legend()
    plt.grid()
    plt.show()

    # Gráfico do erro absoluto
    plt.figure(figsize=(8,5))
    plt.plot(iteracoes_pf, erros_pf, marker='o', linestyle='-', label='Erro Ponto Fixo')
    plt.plot(iteracoes_nr, erros_nr, marker='o', linestyle='-', label='Erro NR')
    plt.xlabel('Iterações')
    plt.ylabel('Erro Absoluto')
    plt.title('Erro Absoluto ao Longo das Iterações')
    plt.yscale('log')  # Escala logarítmica para melhor visualização
    plt.legend()
    plt.grid()
    plt.show()

```

```python
import math

# -----------------------------
# 1) CONSTANTES E PARÂMETROS
# -----------------------------
n = 1000            # mol
p = 3.5e7             # Pa (350 bar)
T = 300.0             # K
R = 8.31446261815324  # J/(mol*K)


a = 0.401             # Pa m^6 / mol^2
b = 42.7e-6           # m^3 / mol

def f_vdw_mol(V): return (p + a*(n/V)**2)*(V - n*b) - n*R*T  # Van der Waals eq

def df_vdw_mol(V):
    alpha = p + a*(n/V)**2; beta = V - n*b
    return -2*a*(n**2)/(V**3)*beta + alpha  # derivada

def bissecao(f, xmin, xmax, tol=1e-12, max_iter=1000):  # raiz se f muda de sinal
    fmin, fmax = f(xmin), f(xmax)
    if fmin*fmax > 0: raise ValueError("Sem mudança de sinal.")
    for i in range(1, max_iter+1):
        xmid = 0.5*(xmin + xmax); fmid = f(xmid)
        if abs(fmid) < tol or 0.5*(xmax - xmin) < tol: return xmid, i
        if fmin*fmid < 0: xmax, fmax = xmid, fmid
        else: xmin, fmin = xmid, fmid
    return 0.5*(xmin + xmax), max_iter

def falsa_posicao(f, xmin, xmax, tol=1e-12, max_iter=1000):  # regula falsi
    fmin, fmax = f(xmin), f(xmax)
    if fmin*fmax > 0: raise ValueError("Sem mudança de sinal.")
    xnew = xmin
    for i in range(1, max_iter+1):
        xnew = xmax - fmax*(xmin - xmax)/(fmin - fmax)
        fnew = f(xnew)
        if abs(fnew) < tol: return xnew, i
        if fmin*fnew < 0: xmax, fmax = xnew, fnew
        else: xmin, fmin = xnew, fnew
    return xnew, max_iter

def newton_raphson(f, df, x0, tol=1e-12, max_iter=100):  # raiz via derivada
    x = x0
    for i in range(1, max_iter+1):
        fx = f(x)
        if abs(fx) < tol: return x, i
        dfx = df(x)
        if abs(dfx) < 1e-30: raise ZeroDivisionError("Derivada quase zero.")
        xn = x - fx/dfx
        if xn <= 0: xn = x*0.5
        x = xn
    return x, max_iter

if __name__ == "__main__":
    vmin, vmax, tol = 1e-4, 1, 1e-20

    print("--- Bissecao ---")
    try:
        vb, ib = bissecao(f_vdw_mol, vmin, vmax, tol, 2000)
        print(f"V={vb}, f(V)={f_vdw_mol(vb)}, it={ib}")
    except ValueError as e:
        print(e)

    print("\n--- Falsa Posição ---")
    try:
        vf, ifp = falsa_posicao(f_vdw_mol, vmin, vmax, tol, 2500000)
        print(f"V={vf}, f(V)={f_vdw_mol(vf)}, it={ifp}")
    except ValueError as e:
        print(e)

    print("\n--- Newton-Raphson ---")
    x0 = 0.05
    try:
        vn, inr = newton_raphson(f_vdw_mol, df_vdw_mol, x0, tol, 2000)
        print(f"V={vn}, f(V)={f_vdw_mol(vn)}, it={inr}")
    except (ValueError, ZeroDivisionError) as e:
        print(e)
```

```python
import sys
import math

EPS = sys.float_info.epsilon

E = -0.33131# Exemplo de valor negativo para E

def f(x):
    return -(1/x**3) - (1/x**2) - E

def df(x):
    # f(x) = -1/x^3 - 1/x^2 - E
    # derivada: df/dx = 3/x^4 + 2/x^3
    return 3.0/x**4 + 2.0/x**3

def bissecao(f, a, b, max_iter=1000):
    """
    Bisseção com critério de parada:
       |x_{n+1} - x_n| <= EPS * max(1, |x_{n+1}|)
    """
    fa, fb = f(a), f(b)
    if fa*fb > 0:
        raise ValueError("Sem mudança de sinal no intervalo.")

    x_old = None
    for i in range(max_iter):
        m = 0.5*(a + b)
        fm = f(m)

        # Se já temos x_old, checamos o critério

        if x_old is not None:
            if abs(m - x_old) <= EPS*max(1, abs(m)):
                return m, i+1

        x_old = m

        # Continua o método de bisseção
        if fa*fm < 0:
            b, fb = m, fm
        else:
            a, fa = m, fm

    return 0.5*(a+b), max_iter

def falsa_posicao(f, a, b, max_iter=1000):
    """
    Falsa Posição com critério:
       |x_{n+1} - x_n| <= EPS * max(1, |x_{n+1}|)
    """

    fa, fb = f(a), f(b)
    if fa*fb > 0:
        raise ValueError("Sem mudança de sinal no intervalo.")

    x_old = None
    for i in range(max_iter):
        c = b - fb*(a - b)/(fa - fb)
        fc = f(c)

        if x_old is not None:
            if abs(c - x_old) <= EPS*max(1, abs(c)):
                return c, i+1

        x_old = c

        if fa*fc < 0:
            b, fb = c, fc
        else:
            a, fa = c, fc

    return c, max_iter

def newton(f, df, x0, max_iter=100):
    """
    Newton-Raphson com critério:
       |x_{n+1} - x_n| <= EPS * max(1, |x_{n+1}|)
    """
    x = x0
    for i in range(max_iter):
        x_old = x
        fx = f(x_old)
        dfx = df(x_old)
        if abs(dfx) < 1e-30:
            raise ZeroDivisionError("Derivada muito pequena.")
        x_new = x_old - fx/dfx

        if abs(x_new - x_old) <= EPS*max(1, abs(x_new)):
            return x_new, i+1

        x = x_new
    return x, max_iter

if __name__ == "__main__":
    a, b = 1e-4, 10.0

    print(f"Epsilon da máquina = {EPS}")

    # Bisseção
    try:
        rb, ib = bissecao(f, a, b)
        print(f"Bissecao: x={rb}, f(x)={f(rb)}, it={ib}")
    except ValueError as e:
        print("Bissecao falhou:", e)

    # Falsa Posição
    try:
        rf, ifp = falsa_posicao(f, a, b)
        print(f"FalsaPos: x={rf}, f(x)={f(rf)}, it={ifp}")
    except ValueError as e:
        print("FalsaPos falhou:", e)

    # Newton
    x0 = 1.0
    try:
        rn, inr = newton(f, df, x0)
        print(f"Newton:  x={rn}, f(x)={f(rn)}, it={inr}")
    except (ValueError, ZeroDivisionError) as e:
        print("Newton falhou:", e)
```
